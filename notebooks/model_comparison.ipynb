```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the saved DataFrame
df = pd.DataFrame({
    "Model": [
        "Linear Regression", "Decision Tree", "Lasso Regression", "Ridge Regression",
        "Elastic Net Regression", "Random Forest Regressor", "XGBoost Regressor", "KNN Regressor"
    ],
    "RMSE": [50.70, 48.76, 50.93, 51.01, 50.93, 48.44, 48.60, 52.07],
    "MAE": [36.01, 33.56, 36.20, 36.41, 36.20, 31.32, 31.68, 37.65],
    "R² Score": [0.622, 0.557, 0.618, 0.617, 0.618, 0.655, 0.652, 0.600]
})

# Create a folder to save images if it doesn't exist
save_path = "static/visualizations"
os.makedirs(save_path, exist_ok=True)

# Plot RMSE Comparison
plt.figure(figsize=(10, 6))
plt.barh(df["Model"], df["RMSE"], color="skyblue")
plt.xlabel("RMSE")
plt.title("Model Comparison - RMSE")
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.savefig(os.path.join(save_path, "rmse_comparison.png"), bbox_inches="tight")
plt.close()

# Plot MAE Comparison
plt.figure(figsize=(10, 6))
plt.barh(df["Model"], df["MAE"], color="lightcoral")
plt.xlabel("MAE")
plt.title("Model Comparison - MAE")
plt.gca().invert_yaxis()
plt.savefig(os.path.join(save_path, "mae_comparison.png"), bbox_inches="tight")
plt.close()

# Plot R² Score Comparison
plt.figure(figsize=(10, 6))
plt.barh(df["Model"], df["R² Score"], color="seagreen")
plt.xlabel("R² Score")
plt.title("Model Comparison - R² Score")
plt.gca().invert_yaxis()
plt.savefig(os.path.join(save_path, "r2_comparison.png"), bbox_inches="tight")
plt.close()

print(f"Visualizations saved in {save_path}")
```

```python
# Ensure the directory exists
output_dir = "static/visualizations"
os.makedirs(output_dir, exist_ok=True)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 5))  # Wider figure for better spacing
ax.axis('tight')
ax.axis('off')

# Create the table
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Adjust font size and alignment
table.auto_set_font_size(False)
table.set_fontsize(12)

# Set column widths
for j in range(len(df.columns)):
    table.auto_set_column_width(j)

# Increase row height
for i in range(len(df) + 1):  # +1 for header row
    for j in range(len(df.columns)):
        table[i, j].set_height(0.1)  # Adjust row height for better spacing

# Save the table as an image
output_path = os.path.join(output_dir, "performance_matrix.png")
plt.savefig(output_path, bbox_inches='tight', dpi=300)

print(f"Table saved successfully at {output_path}")
```
